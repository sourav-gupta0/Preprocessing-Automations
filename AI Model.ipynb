```python
# AI Model for Data and Image Pre-processing (ETL) with Advanced Statistical Charts

import pandas as pd
import numpy as np
import os
from pathlib import Path
from PIL import Image
import pytesseract
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from openpyxl import load_workbook
from openpyxl.drawing.image import Image as XLImage

# Load Data Function
def load_data(file_path):
    if os.path.isdir(file_path):
        image_texts = [{'Image File': img.name, 'Extracted Text': pytesseract.image_to_string(Image.open(img))}
                       for img in Path(file_path).rglob("*.jpg")]
        return pd.DataFrame(image_texts)
    ext = Path(file_path).suffix.lower()
    return pd.read_csv(file_path) if ext == '.csv' else pd.read_excel(file_path) if ext in ['.xls','.xlsx'] else pd.read_json(file_path)

# Data Pre-processing
def preprocess_data(df):
    df.fillna(df.median(numeric_only=True), inplace=True)
    df.fillna('Unknown', inplace=True)
    for col in df.select_dtypes(include='object'):
        df[col] = LabelEncoder().fit_transform(df[col].astype(str))
    df[df.select_dtypes(include=[np.number]).columns] = StandardScaler().fit_transform(df.select_dtypes(include=[np.number]))
    return df

# Generate Detailed Statistical Charts
def generate_excel_report(df, output_file):
    with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
        df.to_excel(writer, sheet_name='Processed Data', index=False)
        df.describe().T.to_excel(writer, sheet_name='Summary')
        pd.DataFrame(df.isnull().sum(), columns=['Missing Values']).to_excel(writer, sheet_name='Missing Values')
        df.corr().to_excel(writer, sheet_name='Correlation Analysis')

    charts = {
        'correlation_heatmap': sns.heatmap(df.corr(), annot=True, cmap='coolwarm'),
        'boxplot': sns.boxplot(data=df.select_dtypes(include=[np.number])),
        'histogram': df.hist(figsize=(12, 8)),
        'pairplot': sns.pairplot(df, diag_kind='kde')
    }

    wb = load_workbook(output_file)
    ws = wb.create_sheet('Charts')
    for idx, (name, chart) in enumerate(charts.items()):
        plt.figure()
        plt.title(name.replace('_', ' ').title())
        plt.savefig(f'{name}.png')
        ws.add_image(XLImage(f'{name}.png'), f'{chr(65 + idx * 5)}1')
    wb.save(output_file)
    print(f"Excel reports with multiple statistical charts generated: {output_file}")

# Execute Model
file_path = input("Enter file or folder path: ")
output_file = 'advanced_statistical_reports.xlsx'
try:
    df = load_data(file_path)
    df = preprocess_data(df)
    generate_excel_report(df, output_file)
except Exception as e:
    print(f"Error: {e}")
```

```python
# AI Model for Data and Image Pre-processing (ETL) with Comprehensive Statistical Charts and Advanced Analysis

import pandas as pd
import numpy as np
import os
from pathlib import Path
from PIL import Image
import pytesseract
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from openpyxl import load_workbook
from openpyxl.drawing.image import Image as XLImage
from scipy.stats import skew, kurtosis

# Load Data Function
def load_data(file_path):
    if os.path.isdir(file_path):
        image_texts = [{'Image File': img.name, 
                        'Extracted Text': pytesseract.image_to_string(Image.open(img))}
                       for img in Path(file_path).rglob("*.jpg")]
        return pd.DataFrame(image_texts)
    ext = Path(file_path).suffix.lower()
    return pd.read_csv(file_path) if ext == '.csv' else pd.read_excel(file_path) if ext in ['.xls','.xlsx'] else pd.read_json(file_path)

# Data Pre-processing
def preprocess_data(df):
    df.fillna(df.median(numeric_only=True), inplace=True)
    df.fillna('Unknown', inplace=True)
    for col in df.select_dtypes(include='object'):
        df[col] = LabelEncoder().fit_transform(df[col].astype(str))
    df[df.select_dtypes(include=[np.number]).columns] = StandardScaler().fit_transform(df.select_dtypes(include=[np.number]))
    return df

# Generate Detailed Excel Report with Additional Charts and Stats
def generate_excel_report(df, output_file):
    # Compute skewness and kurtosis for numerical features
    df_skew = df.select_dtypes(include=[np.number]).apply(skew)
    df_kurt = df.select_dtypes(include=[np.number]).apply(kurtosis)
    
    with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
        df.to_excel(writer, sheet_name='Processed Data', index=False)
        df.describe().T.to_excel(writer, sheet_name='Summary')
        pd.DataFrame(df.isnull().sum(), columns=['Missing Values']).to_excel(writer, sheet_name='Missing Values')
        df.corr().to_excel(writer, sheet_name='Correlation Analysis')
        pd.concat([df_skew, df_kurt], axis=1, keys=['Skewness', 'Kurtosis']).to_excel(writer, sheet_name='Skewness and Kurtosis')
    
    # Create Boxplot of Numerical Features
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=df.select_dtypes(include=[np.number])).set_title('Boxplot of Numerical Features')
    plt.savefig('boxplot.png')
    plt.clf()
    
    # Create Histograms of Numerical Features
    df.hist(figsize=(14, 10), bins=20, edgecolor='black')
    plt.suptitle('Histograms of Numerical Features')
    plt.savefig('histogram.png')
    plt.clf()
    
    # Create Pairplot of Numerical Features
    sns.pairplot(df.select_dtypes(include=[np.number]), diag_kind='kde')
    plt.savefig('pairplot.png')
    plt.clf()
    
    # Create Heatmap of Correlations
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Heatmap of Correlations')
    plt.savefig('heatmap.png')
    plt.clf()
    
    # Additional Visualization: Violin Plot for Numerical Features
    plt.figure(figsize=(14, 8))
    numeric_data = df.select_dtypes(include=[np.number])
    numeric_melt = numeric_data.melt(var_name='Variable', value_name='Value')
    sns.violinplot(x='Variable', y='Value', data=numeric_melt)
    plt.title("Violin Plot of Numerical Features")
    plt.savefig('violin_plot.png')
    plt.clf()
    
    # Additional Visualization: Density Plot (KDE) for Numerical Features
    plt.figure(figsize=(14, 8))
    for col in numeric_data.columns:
        sns.kdeplot(numeric_data[col], shade=True, label=col)
    plt.title("Density Plot of Numerical Features")
    plt.legend()
    plt.savefig('density_plot.png')
    plt.clf()
    
    # Add images to Excel in a "Charts" sheet
    wb = load_workbook(output_file)
    ws = wb.create_sheet('Charts')
    chart_positions = {
        'boxplot.png': 'A1',
        'histogram.png': 'E1',
        'pairplot.png': 'I1',
        'heatmap.png': 'M1',
        'violin_plot.png': 'A20',
        'density_plot.png': 'E20'
    }
    for img_file, cell in chart_positions.items():
        ws.add_image(XLImage(img_file), cell)
    wb.save(output_file)
    print(f"Comprehensive Excel report with advanced charts and statistical analysis created: {output_file}")

# Execute Model
file_path = input("Enter file or folder path: ")
output_file = 'advanced_statistical_reports.xlsx'
try:
    df = load_data(file_path)
    df = preprocess_data(df)
    generate_excel_report(df, output_file)
except Exception as e:
    print(f"Error: {e}")
```
